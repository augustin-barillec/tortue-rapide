```python
# get model
from tensorflow.python.keras.models import load_model, Model
import utils
from matplotlib import pyplot as plt
import numpy as np
from vis.visualization import visualize_cam, visualize_activation
import keras.backend as K
import tensorflow as tf 
import cv2

#load model and images
model = load_model('/home/ysance.local/r.carmona-hagelsteen/Téléchargements/ironcar_second_try.hdf5')

K.set_session(tf.Session(graph=model.output.graph)) 
init = K.tf.global_variables_initializer() 
tf_session = K.get_session().run(init)

X, Y = utils.newtub_to_array('/home/ysance.local/r.carmona-hagelsteen/Téléchargements/set_1/')
```

```python
# conv idx : 2, 3, 4, 5
# 0 : ?
# 1 : right ? 
# 2 : straight ? lower surface

layer_idx = 5
pred_index = 0
im_idx=34

output = model.output[:, pred_index]

layer = model.layers[layer_idx]
grads = K.gradients(output, layer.output)[0]
pooled_grads = K.mean(grads, axis=(0, 1, 2))
iterate = K.function([model.input], [pooled_grads, layer.output[0]])

n=im_idx # image index
img = X[n][30:, :, :]
plt.imshow(X[n])
```

```python
model.predict(X[n].reshape((1,)+X[n].shape)/255-.5)
```

```python
pooled_grads_value, conv_layer_output_value = iterate([X[n].reshape((1,)+X[n].shape)/255-.5])

for i in range(len(pooled_grads_value)):
    conv_layer_output_value[:, :, i] *= pooled_grads_value[i]

heatmap = np.mean(conv_layer_output_value, axis=-1)
heatmap = np.maximum(heatmap, 0)
heatmap /= np.max(heatmap)
heatmap = cv2.resize(heatmap, (160, 80))*255
heatmap = heatmap.astype(np.uint8)
plt.matshow(heatmap)
plt.title('pred : ' + str(pred_index))
```

```python
cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

gray = cv2.cvtColor(X[n][30:110, :, :], cv2.COLOR_RGB2GRAY)
fin = cv2.addWeighted(heatmap, .6, gray, .4, 0)
plt.imshow(fin)
```
